## A General Purpose Car Script

Select SUV and select neighborhood

```python
import time
import random
from datetime import datetime, timedelta
from typing import List, Tuple

# Basic grid imports
from grid.robot.wheeled.airgen_car import AirGenCar
import airgen
from airgen.utils.collect import data_collector
from airgen import WeatherParameter, Vector3r

# Import AI models
from grid.model.perception.detection.gdino import GroundingDINO
from grid.model.perception.segmentation.gsam2 import GSAM2

# initialise the robot
airgen_car_0 = AirGenCar()

# Load AI models
detection = GroundingDINO()
segmentation = GSAM2()

# Load car
client = airgen_car_0.client
```

### Set and Randomize the Weather

```python
weather_types = {
    "Fog": WeatherParameter.Fog,
    "Rain": WeatherParameter.Rain,
    "Snow": WeatherParameter.Snow,
    "Dust": WeatherParameter.Dust
}

# choose a readable option
rand_weather_choice_str = random.choice(list(weather_types.keys()))

# get the numeric value
rand_weather_choice_int = weather_types[rand_weather_choice_str]

client.simEnableWeather(True)
# Adds a random weather type to the scene with a variable intensity from 0 to 1
rand_weather_intensity = random.uniform(0, 1)

client.simSetWeatherParameter(rand_weather_choice_int,rand_weather_intensity)

# Adds a random datetime to the scene in the form ("2024-07-22 17:00:00")
start = datetime(2024, 1, 1)
random_date = start + timedelta(days=random.randint(0, 365), hours=random.randint(0, 23), minutes=random.randint(0, 59), seconds=random.randint(0, 59))
random_datetime_str = random_date.strftime("%Y-%m-%d %H:%M:%S")
client.simSetTimeOfDay(True, random_datetime_str)

print(random_datetime_str)
print(rand_weather_choice_str, rand_weather_intensity)
```
